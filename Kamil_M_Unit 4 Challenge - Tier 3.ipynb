**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Here we create the function to compare the average price of a borough in 1998 and 2018.

def compare_price(input_Borough):
    filter_df = df_copy[df_copy['Borough'] == input_Borough]
    
    avg_98 = filter_df[filter_df['Year'] == 1998]['Avg_price'].mean()
    avg_18 = filter_df[filter_df['Year'] == 2018]['Avg_price'].mean()
    
    ratio = avg_98/avg_18
    
    return ratio

# We can also look to calc the percentage increase
def pct_increase(input_Borough):
    filter_df = df_copy[df_copy['Borough'] == input_Borough]
    
    start = filter_df[filter_df['Year'] == 1998]['Avg_price'].mean()
    end = filter_df[filter_df['Year'] == 2018]['Avg_price'].mean()
    
    diff = ((end - start)/abs(start))*100
    return diff
```

```python
# Remember we already have a list of the unique London Boroughs called london_boroughs
london_boroughs
```

```python
# Now we can filter through all of the London Boroughs and extract the ratios
ratios = {}
pct_diffs = {}
for i in london_boroughs:
    ratios[i] = compare_price(i)
    pct_diffs[i] = pct_increase(i)
```

```python
pct_diffs
```

```python
# Now we can set the ratios per Borough as a dataframe 
final = pd.DataFrame(ratios, index = [0])
final_T = final.transpose().reset_index()
df = final_T.rename(columns={'index': 'Borough', 0:'Ratio'}) # We rename the dataframe to df to make it easier
df['Pct increase'] = pct_diffs.values()
df.sort_values(by='Pct increase', ascending=False)
```

```python
# Finally we can take a look at the top 10 and plot
top_10 = df.sort_values(by='Pct increase', ascending=False)[:10]
top_10
```

```python
ax = top_10[['Borough', 'Pct increase']].plot(kind='bar')
ax.set_xticklabels(top_10['Borough'])
ax.set_ylabel('Percentage Increase')
ax.get_legend().remove()
ax.set_ylim((0,600))
ax.set_title('London Boroughs percentage of house price increase 1998-2018')
```

```python
print('Average Percent increase for all Longon Boroughs: ', df['Pct increase'].mean())
print('Average Percent increase for the top 10 Longon Boroughs: ', top_10['Pct increase'].mean())
```
